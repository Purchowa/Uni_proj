Zadanie 1 - wczytanie zbioru danych
Target w zbiorze danych jest używany do uczenia nadzorowanego informuje, że dany przykład należy do danej klasy. W tym przypadku jaki to kwiatek.

```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

print(df.head())
```

```python
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
df = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
df['target'] = breast_cancer.target

print(df.head())
```

Zadanie 2 - oczyszczanie danych
Wnioski: nie ma brakujących danych.

```python
missing_data = df.isna().sum()
print(f"Brakujace dane:\n{missing_data}")
df_cleaned = df.dropna()
```

Zadanie 3 - wizualizacja
W przypadku cech sepal length oraz sepal width dla przypadku dwóch kwiatów w obszarze (3, 6) jest ciężko je odróżnić.

```python
from matplotlib import pyplot as plt
FEATURE_1 = 'mean radius'
FEATURE_2 = 'mean texture'

plt.figure(figsize=(8, 6))
for species in df_cleaned['target'].unique(): # for each species [0, 1, ..., n]
    subset = df_cleaned[df_cleaned['target'] == species] # create subset of one specie
    plt.scatter(subset[FEATURE_1], subset[FEATURE_2], label = iris.target_names[species])

plt.xlabel(FEATURE_1)
plt.ylabel(FEATURE_2);
```

Zadanie 4 - Analiza liczebności klas
Wszystkie klasy są równoliczne

```python
df_number_of_samples = df_cleaned['target'].value_counts()
print(f"Liczba próbek dla każdej z klas\n{df_number_of_samples}")
```

Zadanie 5 - histogram

```python
def histogram_for_feature(feature: str):
    plt.figure(figsize=(8, 6))
    plt.hist(df[feature], bins=20, edgecolor='k')
    plt.xlabel(feature)
    plt.ylabel("Liczba probek")
    plt.title(f"Histogram {feature}")
    plt.show()
```

```python
histogram_for_feature(FEATURE_1)
```

Histogram dla klas to trzy słupki, bo tyle mamy klas.

```python
histogram_for_feature("target")
```
